```python
# Correlation Heatmap exclude Label

import seaborn as sns
import matplotlib.pyplot as plt

# Select only numerical features for correlation calculation
numerical_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(10,6))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
if 'label' in df.columns:
    df["crop"] = df["label"]  # Assign 'label' to crop_column
else:
    df["crop"] = df["Crop"] # keep original name if it is correct
```

```python
# Crop Distribution

plt.figure(figsize=(12,6))
sns.countplot(y=df['crop'], order=df['crop'].value_counts().index, palette='viridis')
plt.title('Distribution of Crops in the Dataset')
plt.show()
```

```python
# Impact of Rainfall on Crop Yield

sns.boxplot(x=df['crop'], y=df['rainfall'])
plt.xticks(rotation=90)
plt.title('Rainfall vs. Crop Type')
plt.show()
```
