# ***2. EDA(Exploratory Data Analysis)***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned Excel file
file_path = '/content/Cleaned_Workforce_Data.xlsx'
data = pd.read_excel(file_path)
```

```python
# Descriptive statistics for numerical columns
print(data.describe())

# Summary for categorical columns
print(data.describe(include=['object']))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Attrition Count
attrition_counts = data['Attrition'].value_counts()

# Bar Plot
plt.figure(figsize=(8, 5))
sns.barplot(x=attrition_counts.index, y=attrition_counts.values, palette='viridis')
plt.title('Attrition Distribution')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Attrition', hue='Gender', data=data, palette='Set2')
plt.title('Attrition by Gender')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='CF_age band', hue='Attrition', data=data)
plt.title('Attrition by Age Band')
plt.xlabel('Age Band')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Department', hue='Attrition', data=data, palette='husl')
plt.title('Attrition by Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.show()
```

```python
# Check for numeric columns
numeric_cols = data.select_dtypes(include='number')

# Correlation Matrix
correlation_matrix = numeric_cols.corr()

# Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Group data by Job Role and Attrition, then count the occurrences
attrition_by_role = data.groupby(['Job Role', 'Attrition']).size().unstack().fillna(0)

# Create a bar plot
plt.figure(figsize=(12, 6))
attrition_by_role['Yes'].sort_values(ascending=False).plot(kind='bar', color='lightcoral')

# Add titles and labels
plt.title('Highest Attrition by Job Roles')
plt.xlabel('Job Role')
plt.ylabel('Attrition Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Count the number of employees in each age group
age_group_counts = data['CF_age band'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(age_group_counts, labels=age_group_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set3', len(age_group_counts)))
plt.title('Distribution of Employees Across Age Groups')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Count the number of employees in each age group
age_group_counts = data['CF_age band'].value_counts()

# Create a donut chart
plt.figure(figsize=(8, 8))

# Plotting the donut chart
plt.pie(age_group_counts, labels=age_group_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set2", len(age_group_counts)))

# Draw a circle in the middle to make it a donut chart
centre_circle = plt.Circle((0,0), 0.60, color='white', fc='white', linewidth=1.25)
plt.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Employee Distribution Across Age Groups')
plt.tight_layout()

# Show the plot
plt.show()
```
